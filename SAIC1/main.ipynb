```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("cw2dane.csv")
df1["Ku"] = df1["Uwypp[V]"]/df1["Uwepp[V]"]
df1
```

```python
plt.plot(df1["f[Hz]"].to_numpy(), df1["Ku"].to_numpy(), 'r')
plt.scatter(df1["f[Hz]"].to_numpy(), df1["Ku"].to_numpy())
plt.grid(True)
plt.xscale('log')
plt.show()
```
